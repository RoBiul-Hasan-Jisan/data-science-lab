# Phase 7: Outlier Detection

An outlier is a data point that significantly differs from other observations.

*Outliers can:*

- Affect model performance.
- Distort statistical measures such as mean and standard deviation.
- Increase model error.

## 1. Detect Outliers Using IQR Method

The Interquartile Range (IQR) method is one of the most commonly used techniques.

IQR = Q3 - Q1

Lower Bound:

Q1 - 1.5 × IQR

Upper Bound:

Q3 + 1.5 × IQR

````python
def detect_outliers_iqr(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[
        (df[column] < lower_bound) |
        (df[column] > upper_bound)
    ]

    return len(outliers)

for col in df.columns:
    print(col, detect_outliers_iqr(col))
````

## Detect Outliers in All Columns

````python
for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    n_outliers = ((df[col] < lower) |
                  (df[col] > upper)).sum()

    print(f"{col}: {n_outliers}")
````

## 2. Detect Outliers Using Z-Score

The Z-score measures how many standard deviations a point is away from the mean.

**5. How Would You Treat Outliers?**

This is a very common interview question.

*Method 1: Remove Outliers*
```
Q1 = df['population'].quantile(0.25)
Q3 = df['population'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df_clean = df[
    (df['population'] >= lower) &
    (df['population'] <= upper)
]
```
*Use when:*
- Outliers are data entry errors.
- Very few outliers exist.

*Method 2: Capping (Winsorization)*

Replace extreme values with boundary values.
```
lower = df['population'].quantile(0.01)
upper = df['population'].quantile(0.99)

df['population'] = df['population'].clip(lower, upper)
```

*Method 3: Transformation*

Apply log transformation.
```
import numpy as np

df['population_log'] = np.log1p(df['population'])
```
Useful for highly skewed distributions.

*Method 4: Use Robust Models*

Use algorithms less affected by outliers:

- Random Forest
- XGBoost
- Decision Trees
- Advanced Interview Questions

## 3. Explain Multicollinearity

> Multicollinearity occurs when two or more independent features are highly correlated with each other.

For example:

```total_rooms ↔ total_bedrooms```

If correlation is extremely high:

```|correlation| > 0.8```

we may suspect multicollinearity.

Common Rule

```
|r| > 0.8
```
or
```
|r| > 0.9
```
often indicates strong multicollinearity.

Example

Suppose:
```
total_rooms = 3000
total_bedrooms = 700
```
These features naturally increase together.

Hence:
```
total_rooms ↑
total_bedrooms ↑
```
This creates multicollinearity.

## 5. How Would You Handle Multicollinearity?

- Method 1: Drop Highly Correlated Features

Example:

```df = df.drop('total_bedrooms', axis=1)```

if:

```corr(total_rooms, total_bedrooms) > 0.9```

- Method 2: Feature Engineering

Instead of:
```
total_rooms
total_bedrooms
```
Create:
```
df['bedrooms_per_room'] = (
    df['total_bedrooms'] /
    df['total_rooms']
)
```
and remove one of the original features.

- Method 3: Variance Inflation Factor (VIF)
```
VIF quantitatively measures multicollinearity.

from statsmodels.stats.outliers_influence \
import variance_inflation_factor

X = df.drop('median_house_value', axis=1)

vif = pd.DataFrame()
vif['Feature'] = X.columns

vif['VIF'] = [
    variance_inflation_factor(X.values, i)
    for i in range(X.shape[1])
]

print(vif)

```

| VIF   | Interpretation       |
| ----- | -------------------- |
| 1     | No multicollinearity |
| 1 - 5 | Moderate             |
| > 5   | High                 |
| > 10  | Severe               |

**What is VIF?**


> Variance Inflation Factor (VIF) measures how much the variance of a regression coefficient is inflated due to multicollinearity.

**Assumptions of Linear Regression**

> This is one of the most frequently asked interview questions.

1. Linearity

There should be a linear relationship between independent variables and the target variable.

Example:
```
sns.scatterplot(
    x='median_income',
    y='median_house_value',
    data=df
)
```
2. Independence of Errors

> Residuals (errors) should be independent of each other.

3. No Multicollinearity

> Independent variables should not be highly correlated with each other.

Check using:
```
df.corr()

or

VIF
```
4. Homoscedasticity

> The variance of residuals should remain constant across all predicted values.

5. Normality of Residuals

> Residuals should approximately follow a normal distribution.

6. No Significant Outliers

> Extreme outliers can disproportionately influence the regression line.



**What happens if assumptions are violated?**
| Assumption Violated  | Possible Effect                  |
| -------------------- | -------------------------------- |
| Non-linearity        | Poor predictions                 |
| Multicollinearity    | Unstable coefficients            |
| Heteroscedasticity   | Unreliable statistical inference |
| Non-normal residuals | Invalid confidence intervals     |
| Outliers             | Distorted model                  |

Using:

- Residual plots

# Phase 12: Model Evaluation

> After building a model, we need to evaluate how well it performs on unseen data.

For regression problems like house price prediction, the most commonly used metrics are:

- MAE (Mean Absolute Error)
- MSE (Mean Squared Error)
- RMSE (Root Mean Squared Error)
- R² Score

## 1. Import Evaluation Metrics

````python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

import numpy as np
````

## 2. Calculate MAE

````python
mae = mean_absolute_error(y_test, y_pred)

print("MAE:", mae)
````

> MAE measures the average absolute difference between actual and predicted values.

## 3. Calculate MSE

````python
mse = mean_squared_error(y_test, y_pred)

print("MSE:", mse)
````

> MSE measures the average squared error between actual and predicted values.

## 4. Calculate RMSE

````python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("RMSE:", rmse)
````

## 5. Calculate R² Score

````python
r2 = r2_score(y_test, y_pred)

print("R² Score:", r2)
````

**Which metric is most important for house price prediction?**

> RMSE is often considered the most important metric for house price prediction because it penalizes large prediction errors more heavily and is expressed in the same units as house prices.

**Comparison of Metrics**
| Metric | Interpretation                  | Sensitive to Outliers |
| ------ | ------------------------------- | --------------------- |
| MAE    | Average absolute error          | No                    |
| MSE    | Average squared error           | Yes                   |
| RMSE   | Average error in original units | Yes                   |
| R²     | Explained variance              | No                    |


> For the California Housing dataset, I would primarily focus on RMSE because house price prediction requires minimizing large prediction errors. I would also report MAE and R² Score to provide a complete assessment of model performance.
